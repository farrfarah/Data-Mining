## Transformasi data Label Encoding

```python
df = pd.read_csv('/content/student.csv', header = 0) # Membaca file csv, baris pertama dimulai dari 0
df.head()
```

```python
 df = df.dropna() # Menghapus semua missing value
```

```python
datakolom = df.select_dtypes(exclude=[np.number]).columns # Membuat df baru berisi kolom non-numerik dengan mengecualikan kolom numerik menggunakan library numpy
datakolom
```

```python
 df[datakolom].head() # Menampilkan 5 baris pertama dataframe yg berisi kolom non numerik
```

```python
from sklearn.preprocessing import LabelEncoder # Mengimport LabelEncoder dari modul preprocessing dalam library Scikit-Learn

label_encoder = LabelEncoder() # Membuat objek instance dari LabelEncoder

for i in datakolom: # Perulangan i pada data kolom / melakukan iterasi
  df[i] = label_encoder.fit_transform(df[i]) # Di lakukan transformasi menggunakan objek instance sebelumnya

print("Label Encode Data:") # Output
df.head() # Menampilkan 5 baris pertama data frame
```

```python
 # Simpan data baru
df.to_csv("data_Label_encoding.csv")
```

## Analisis :  
Pada proses label encoding ini saya menggunakan data student.csv. Dalam tahap ini dilakukan penghapusan missing value terlebih dahulu. Selanjutnya, pemisahan data non numerik, pada dataset yang digunakan didapatkan 3 kolom non numerik yaitu Gender, Grade, dan Employed yang dijadikan satu menjadi satu df. Untuk melakukan label encoding maka perlu LabelEncoder dari modul preprocessing dalam Library Scikit-Learn. Lalu, dilakukan transformasi data non numerik dengan perulangan atau iterasi menjaddi data numerik.

## d) Dari semua dataset yang ada di praktikum kali ini pililhlah minimal satu dataset kemudian lakukan Data Discretization!. (poin : 20)

```python
dataset = "/content/mark.csv" # Definisi dataset mark menjadi dataset
df = pd.read_csv(dataset, header = 0) # Membaca file csv dataset dengan dimulai dari 0
```

```python
df.head() # Menampilkan 5 dataset pertama
```

```python
# Lakukan pengelompokan menggunakan fungsi pd.cut() pada kolom marks dan tampilkan 10 kolom teratas.
df['bucket'] = pd.cut(df['Mark'],3, labels=['Poor', 'Average', 'Excellent']) #Melakukan pengelompokan dengan 3 variabel yaitu poor, average dan excellent pada mark
df.head(10) # Menampikan 10 baris pertama
```

```python
# Simpan data baru
df.to_csv("data_discretization.csv")
```

## Analisis :
Pada data discretization ini saya menggunakan dataset mark.csv yang terdapat 3 kolom yaitu student_id, mark dan city. Dalam tahap ini, akan dilakukan pengelompokan fungsi pada kolom Mark dengan memberi 3 label pada nilai mark menggunakan fungsi pd.cut(). Fungsi pd.cut() merupakan fungsi yang terdapat pada library pandas digunakan untuk mengelompokkan data numerik menjadi  ategori atau interval yang lebih mudah diinterpretasikan atau
